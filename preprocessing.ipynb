```python
# Table 1: OWNER info
owner_info = dataset[["Owner1", "Previous Owner", "Mail3", "Mail4"]]

# Table 2: House info
address_info = dataset[["House Number", "Street", "Zipcode","SBL"]]

#Table 3: Zipcode table
zipcode_info = dataset[["Zipcode","State","City"]]

# Table 4: Deed info
deed_info = dataset[["Deed Book", "Deed Page", "Deed Date", "Roll"]]

#table 5:Census table
census_info =  dataset[["Census Tract"]]

# Table 6: Property details
property_info = dataset[["SBL", "Front", "Depth", "Property Class", "Total Living Area"]]

#Table 7: Property class table
property_class_info = dataset[["Property Class","Property Class Description"]]

# Table 8: Price details
price_info = dataset[["Land Value", "Total Value", "Sale Price"]]

# Generating surrogate primary keys
owner_info['Owner_ID'] = dataset["Index"]
deed_info['Deed_ID'] = dataset["Index"]
price_info['Price_ID'] = dataset["Index"]
census_info['Census_ID'] = dataset["Index"]

#Table 9: link table
link_table = pd.DataFrame({
    "SBL": property_info["SBL"],
    "Owner_ID": owner_info["Owner_ID"],
    "Deed_ID": deed_info["Deed_ID"],
    "Price_ID": price_info["Price_ID"],
    "Census_ID": census_info["Census_ID"]
})
```
